```python
# Importa bibliotecas do Python para manipulação e visualização de dados
import pandas as pd # para manipulação
import plotly.express as px  # para visualização

# Definição das atividades
dados = {
    "Atividade": ["Reunião", "Planejamento", "Aquisições", "Parte 1", "Parte 2"],
    "Início": ["2024-06-29", "2024-07-03", "2024-07-05", "2024-07-10", "2024-07-20"],
    "Duração": [1, 2, 5, 15, 20]
}

# Criar DataFrame
df = pd.DataFrame(dados)

# Converter coluna de início para datetime
df["Início"] = pd.to_datetime(df["Início"])

# Calcular término (início + duração)
df["Término"] = df["Início"] + pd.to_timedelta(df["Duração"], unit="D")

# Criar gráfico de Gantt
fig = px.timeline(df, x_start="Início", x_end="Término", y="Atividade", color="Atividade")

# Inverter ordem das tarefas no eixo Y
fig.update_yaxes(autorange="reversed")

# Exibir gráfico
fig.show()

# Instalar Chrome para o Kaleido
!plotly_get_chrome

# Salvar gráfico como PNG (veja a saída na pasta de arquivos do Colab)
fig.write_image("cronograma_gantt.png")
```
